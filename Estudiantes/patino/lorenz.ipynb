encontrar el x,y,z de las funciones de lorenzo

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly as py
from scipy.integrate import odeint
import plotly.graph_objects as go
```

\begin{equation}
\frac{\mathrm{d}x}{\mathrm{d}t} = \sigma (y-x)
\end{equation}


\begin{equation}
\frac{\mathrm{d}y}{\mathrm{d}t} = x(\rho-z)-y
\end{equation}

\begin{equation}
\frac{\mathrm{d}z}{\mathrm{d}t} = xy-\beta z
\end{equation}

```python
def func(r, t, alpha, beta, p):
    x, y, z = r
    dxdt = alpha * (y - x)
    dydt = x * (p - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# =======================
# Condiciones iniciales
# =======================
beta = 8/3  #para realizar los siguientes calculos solamente toca cmabiar las condiciones iniciales
alpha = 10
p = 28 
y0 = [1, 1, 1]
# ============================
# Fin De Condiciones Iniciales
# ============================

t = np.linspace(0, 25, 10000)
sol = odeint(func, y0, t, args=(alpha, beta, p))
x, y, z = sol.T
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=x, y=y, z=z, mode='lines', line=dict(color='blue', width=2)))

# Configurar el diseo del grfico
fig.update_layout(title='Atratctor de lorenz Lorenz',
                  scene=dict(xaxis_title='X',
                             yaxis_title='Y',
                             zaxis_title='Z'),
                  margin=dict(l=0, r=0, b=0, t=40))

# Mostrar el grfico
fig.show()
```
