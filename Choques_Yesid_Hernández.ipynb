```python
import numpy as np
import matplotlib.pyplot as plt

def calculo_vfinal(m1,m2,v1,u1):
  vf1=((m1-m2)*v1+2*m2*u1)/(m1+m2)
  vf2=((m2-m1)*u1+2*m1*v1)/(m1+m2)
  return vf1,vf2

def calculo_choques(m1,m2,v,u):
#U <v y m1<m2
  n=0
  while True:
    v2,u2=calculo_vfinal(m1,m2,v,u)
    if v2<0:
      n=n+1
      v2=-1*v2
    v=v2
    u=u2
    n=n+1
    if u>v and (v and u) >0:
      break
  return n

calculo_choques(1,10000000000,0,-8)
```
